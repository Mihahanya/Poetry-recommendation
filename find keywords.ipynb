```python
from os import walk
from tqdm import tqdm


def text_from_dir(path):
    data_text = ''
    for dirpath, dirnames, filenames in walk(path):
        for file_name in tqdm(filenames):
            if file_name[-4:] != '.txt': continue

            current_text = open(dirpath + '/' + file_name, 'rb').read().decode(errors='replace')
            data_text += current_text

    print('final len', len(data_text))

    return data_text

articles_text = text_from_dir('E:/databases/статьи о стихах/')
common_text = text_from_dir('E:/databases/Russian texts database/')
```

```python
from pymorphy2 import MorphAnalyzer
from collections import Counter
import re  

morph = MorphAnalyzer(lang='ru')

def freq_distribution(text):
	words_seq = re.findall(r'[А-я]+', text.lower())
	lemmatized_words = [morph.parse(word)[0].normal_form for word in tqdm(words_seq)]

	word_freq = {w: 0 for w in set(lemmatized_words)}
	for w in tqdm(lemmatized_words):
		word_freq[w] += 1
	
	word_freq = {k: v/len(lemmatized_words) for k, v in word_freq.items()}

	return word_freq

article_freq = freq_distribution(articles_text)
common_freq = freq_distribution(common_text[:len(articles_text)])
```

```python
norm_article_freq = article_freq
for w, c in article_freq.items():
	if w in common_freq: norm_article_freq[w] -= common_freq[w]

norm_article_freq = [(k, v) for k, v in norm_article_freq.items()]
norm_article_freq.sort(key=lambda a: a[1], reverse=True)
```

```python
for word, freq in norm_article_freq[:300]:
	print(word, freq)
```
